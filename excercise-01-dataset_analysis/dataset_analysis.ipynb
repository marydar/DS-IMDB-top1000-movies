# DataScience excercise-01-Dataset Analysis

### DATASET : IMDB TOP 1000 MOVIES

```python
import pandas as pd
import numpy as np
```

#### 00-Load Data

```python
df = pd.read_csv("archive/imdb_top_1000.csv")
```

#### 01-DATASET Dimensions

```python
print("DATASET DIMENSIONS\n")
print(f"Dataset Shape:\n {df.shape[0]} rows (movies) x {df.shape[1]} columns(features)")
print(f"\nfeatures : ")
print(df.columns)
```

```python
df.describe()
```

#### 03-Detailed Column Statistics

```python
print("=" * 50)
print("COLUMNS")
print("=" * 50)

for col in df.columns:
    # Get basic info
    dtype_info = df[col].dtype
    unique_count = df[col].nunique()
    null_count = df[col].isnull().sum()
    null_pct = (null_count / len(df)) * 100 if len(df) > 0 else 0
    
    print(f"\nColumn Name: {col}")
    print(f"  Data Type: {dtype_info}")
    print(f"  Unique Values Count: {unique_count}")
    print(f"  Missing Values: {null_count} ({null_pct:.2f}%)")
```

#### 05- Data Type Conversion (Numeric Fields)

```python
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce').astype('Int64')
```

```python
df['Runtime'] = df['Runtime'].str.replace(' min', '', regex=False)
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')
```

```python
df['Gross'] = df['Gross'].str.replace(',', '', regex=False)
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
```

```python
df.dtypes
```

```python
df.describe()
```

#### 06-Missing Values and NAN Values

```python
print(df.isnull().sum().sort_values(ascending=False))

# Percentage of missing
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.sort_values(ascending=False)
```

#### 07-Ranking by Imdb Rating, Gross, No_of_Votes & Runtime

```python
top_movies = df.sort_values(by='IMDB_Rating', ascending=False).head(10)
top_movies[['Series_Title', 'IMDB_Rating', 'Director']]
```

```python
top_gross = df.sort_values(by='Gross', ascending=False).head(10)
top_gross[['Series_Title', 'Gross']]
```

```python
top_votes = df.sort_values(by='No_of_Votes', ascending=False).head(10)
top_votes[['Series_Title', 'No_of_Votes']]
```

```python
top_runtime = df.sort_values(by='Runtime', ascending=False).head(10)
top_runtime[['Series_Title', 'Runtime']]
```

```python
top_runtime = df.sort_values(by='Runtime', ascending=False).tail(10)
top_runtime[['Series_Title', 'Runtime']]
```
